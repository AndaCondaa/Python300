```python
#35  ->>> % formatting  ->>>>>> %s: 문자열, %d: int형
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print("이름: %s 나이: %d" % (name1, age1))
print("이름: %s 나이: %d" % (name2, age2))
```

```python
#36   ->>>>> .format() 메서드
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print("이름: {이름} 나이: {나이}".format(이름=name1, 나이=age1))
print("이름: {이름} 나이: {나이}".format(이름=name2, 나이=age2))
```

```python
#37   ->>>>>> f-string:f-string은 문자열 앞에 f가 붙은 형태입니다.
#     ->>>>>> f-string을 사용하면 {변수}와 같은 형태로 문자열 사이에 타입과 상관없이 값을 출력할 수 있습니다.
name1 = "김민수"
age1 = 10
name2 = "이철희"
age2 = 13
print(f"이름: {name1} 나이: {age1}")
print(f"이름: {name2} 나이: {age2}")
```

```python
#76 ->>> 튜플의 값은 변경할 수 없기 때문에, 리스트와 달리 아래 코드는 동작하지 않습니다. 
# 새로운 튜플을 만들고 t 라는 변수를 업데이트 해야 합니다. 기존의 튜플 ('a', 'b', 'c')은 자동으로 삭제됩니다.
t = ('a', 'b', 'c')
print(type(t))
t = ('A', 'b', 'c')
print(type(t))
print(t)
```
